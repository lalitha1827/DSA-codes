```python
def palindrome_substring(s):
    n = len(s)

    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = s[i:j]
            reverse_sub = substring[::-1]

            if substring == reverse_sub and len(substring) > 1:
                print(f"The substring '{substring}' is a palindrome.")
                return

    print("No palindrome substrings with more than one character were found.")
s=input("enter the string:")
palindrome_substring(s)
```

```python
def find_all_palindrome_substrings(s):
    n = len(s)
    found_palindromes = False

    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = s[i:j]
            reverse_sub = substring[::-1]

            if substring == reverse_sub and len(substring) > 1:
                print(f"The substring '{substring}' is a palindrome.")
                found_palindromes = True

    if not found_palindromes:
        print("No palindrome substrings with more than one character were found.")

s=input("enter the string:")
find_all_palindrome_substrings(s)
```

# **two pointers sum**

```python
def two_sum_sorted(arr, target):
    # Initialize two pointers
    left, right = 0, len(arr) - 1

    # Loop until pointers cross
    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return (arr[left], arr[right])  # Found pair

        # If current sum is too small, move left pointer right
        if current_sum < target:
            left += 1
        else:
            # If current sum is too large, move right pointer left
            right -= 1

    return None  # No pair found

# Example usage:
arr = [1, 2, 3, 4, 6, 8, 12]
target = 10
print(two_sum_sorted(arr, target))  # Output: (2, 8)
```

# **sliding window**

```python
def max_subarray_sum(arr, k):
    # Step 1: Compute sum of first window of size k
    window_sum = sum(arr[:k])
    max_sum = window_sum

    # Step 2: Slide the window from start to end
    for i in range(k, len(arr)):
        # Remove element going out of window and add the new element
        window_sum = window_sum - arr[i - k] + arr[i]
        max_sum = max(max_sum, window_sum)

    return max_sum

# Example usage:
arr = [2, 1, 5, 1, 3, 2]
k = 3
print(max_subarray_sum(arr, k))  # Output: 9 (5 + 1 + 3)
```

# **detecting cycle in linked list**


```python
class ListNode:
    def __init__(self, val=0):
        self.val = val        # Store the data value of the node
        self.next = None      # Pointer to the next node (None by default)

def hasCycle(head: ListNode) -> bool:
    """
    Detects whether a linked list has a cycle.
    Returns True if there is a cycle, otherwise False.
    """
    slow = head   # Tortoise pointer (moves 1 step)
    fast = head   # Hare pointer (moves 2 steps)

    while fast and fast.next:
        slow = slow.next          # Move slow pointer by 1
        fast = fast.next.next     # Move fast pointer by 2

        if slow == fast:          # If they meet, there is a cycle
            return True

    return False                  # If fast reaches the end, no cycle


# --------- Example Usage ----------
# Create nodes
node1 = ListNode(3)
node2 = ListNode(2)
node3 = ListNode(0)
node4 = ListNode(-4)

# Link nodes to form a cycle: 3 → 2 → 0 → -4 → back to 2
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = None   # Creates the loop

# Detect cycle
print(hasCycle(node1))  # Output: True
```

# **longest common prefix**

```python
from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ""

        prefix = strs[0]

        for word in strs[1:]:
            while word.find(prefix) != 0:
                prefix = prefix[:-1]

                if not prefix:
                    return ""

        return prefix


# --------- TESTING THE FUNCTION WITH GIVEN INPUTS ---------

# Example 1
strs1 = ["flower", "flow", "flight"]
result1 = Solution().longestCommonPrefix(strs1)
print(f"Input: {strs1}\nOutput: \"{result1}\"\n")

# Example 2
strs2 = ["dog", "racecar", "car"]
result2 = Solution().longestCommonPrefix(strs2)
print(f"Input: {strs2}\nOutput: \"{result2}\"\n")
```

# **reversing linked list**

```python
 # Definition for singly-linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    prev = None        # Previous node starts as None
    curr = head        # Current node starts at head
    while curr:        # Loop until the end of the list
        nxt = curr.next  # Save the next node
        curr.next = prev # Reverse the link
        prev = curr      # Move prev forward
        curr = nxt       # Move curr forward
    return prev          # Prev is the new head


# -------- TEST --------
# Create a simple list: 1 -> 2 -> 3
head = ListNode(1, ListNode(2, ListNode(3)))

# Reverse the list
new_head = reverseList(head)

# Print reversed list
while new_head:
    print(new_head.val, end=" -> " if new_head.next else "")
    new_head = new_head.next
```

# **anagram**

```python
def isAnagram(s1, s2):
    # Step 1: If lengths are different, they cannot be anagrams
    if len(s1) != len(s2):
        return False

    # Step 2: Create a frequency dictionary for s1
    freq = {}
    for char in s1:
        freq[char] = freq.get(char, 0) + 1  # Count occurrence of each character

    # Step 3: Reduce the frequency using characters from s2
    for char in s2:
        if char not in freq:
            return False  # Character in s2 not in s1 → not an anagram
        freq[char] -= 1
        if freq[char] < 0:
            return False  # More occurrences in s2 than in s1 → not an anagram

    # Step 4: All counts should be zero now
    return True


# ---------------- TESTING ----------------
print(isAnagram("listen", "silent"))  # True
print(isAnagram("hello", "bello"))    # False
print(isAnagram("anagram", "nagaram"))# True
```

# **binary search**

```python
def binary_search(arr, target):
    left = 0                 # Start index
    right = len(arr) - 1     # End index

    while left <= right:
        mid = (left + right) // 2  # Middle index
        if arr[mid] == target:
            return mid       # Target found at index mid
        elif arr[mid] < target:
            left = mid + 1   # Search in the right half
        else:
            right = mid - 1  # Search in the left half

    return -1  # Target not found


# ---------------- TESTING ----------------
arr = [1, 3, 5, 9, 11, 13]
target = 9
result = binary_search(arr, target)
print(f"Element {target} found at index: {result}")  # Output: 3
```

# **quick sort**

```python
def quick_sort(arr):
    # Base case: if array has 0 or 1 element, it's already sorted
    if len(arr) <= 1:
        return arr

    pivot = arr[0]            # Choose the first element as pivot
    left = [x for x in arr[1:] if x <= pivot]   # Elements less than or equal to pivot
    right = [x for x in arr[1:] if x > pivot]   # Elements greater than pivot

    # Recursively sort left and right, and combine with pivot
    return quick_sort(left) + [pivot] + quick_sort(right)


# ---------------- TESTING ----------------
arr = [10, 7, 8, 9, 1, 5]
sorted_arr = quick_sort(arr)
print("Original array:", arr)
print("Sorted array:  ", sorted_arr)
```

# **subarray sums**

```python
def subarray_sums(arr):
    n = len(arr)
    result = []  # To store sums of all subarrays

    # Loop for starting index
    for start in range(n):
        current_sum = 0
        # Loop for ending index
        for end in range(start, n):
            current_sum += arr[end]  # Add element to current sum
            result.append(current_sum)  # Store sum

    return result

# --------------- TESTING ----------------
arr = [1, 2, 3]
print("All subarray sums:", subarray_sums(arr))
```
